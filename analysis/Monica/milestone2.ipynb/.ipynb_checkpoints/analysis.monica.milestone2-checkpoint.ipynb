```python
#These are the steps to bring the file into pandas, 
#rename the columns so they are less clunky and 
#removed all the nan values.
#This would be what we can use for Step 3 of Task 3
#Feel free to adjust the  names I chose
#I don't think we need to drop any columns. The only columns to add would be if we want wait times in days (right now they are in weeks) or if we want to multiply the # waiting by the wait_med. 
import pandas as pd
import numpy as np
wd =  (
        pd.read_csv('2009_2020_annual_surgical_wait_times.csv')
        .rename(columns={'FISCAL_YEAR':'year', 'HEALTH_AUTHORITY': 'h_a', 'HOSPITAL_NAME':'hosp', 'PROCEDURE_GROUP': 'prcd', 'WAITING':'waiting', 'COMPLETED': 'comp', 'COMPLETED_50TH_PERCENTILE': 'wait_med', 'COMPLETED_90TH_PERCENTILE':'wait_90'})
        .dropna()
)
wd.head()
```

```python
#These are the steps to convert waiting and comp columns from str to int 
#I had to get rid of the commas 
#I changed the <5 people waiting or <5 procedures done arbitrarily over to the value 3 (half way between 1 and 5)
#We just need to make this intoa function and do step3 of task 2...which I think will be easy enough
wd['waiting'] = (
                  wd['waiting'].str.replace(',','')
                                .str.replace('<5','3')
)

wd['comp'] = (
               wd['comp'].str.replace(',','')
                          .str.replace('<5','3')
)

wd['waiting'] = pd.to_numeric(wd['waiting'])
wd['comp'] = pd.to_numeric(wd['comp'])
wd.head()
```

```python
#To demonstrate that the waiting and comp values are now numeric
wd.describe()
```
